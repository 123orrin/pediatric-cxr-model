```python
import pandas as pd
```

```python
# spreadsheet with all annotations
ANNOT_DIR = "/home/orrin/aps360/pediatric-cxr-model/data_cleaned/annotations"

df_kermany = pd.read_csv(f"{ANNOT_DIR}/kermany_annotations.csv")
df_heart = pd.read_csv(f"{ANNOT_DIR}/congenital_heart_disease_annotations.csv")
df_nih = pd.read_csv(f"{ANNOT_DIR}/nih_annotations.csv")
df_pedcxr = pd.read_csv(f"{ANNOT_DIR}/pedcxr_annotations.csv")

df_kermany["origin"] = "kermany"
df_heart["origin"] = "heart"
df_nih["origin"] = "nih"
df_pedcxr["origin"] = "pedcxr"

df_all = pd.concat([df_kermany, df_heart, df_nih, df_pedcxr], ignore_index=True)

def remove_one_hotness(df):
    # get label
    df['label'] = df.apply(lambda row: row[row == 1].index[0] if 1 in row.values else 'no finding', axis=1)
    df = df[['filename', 'origin', 'label', 'has bbox', 'min_x', 'min_y', 'width', 'height']]
    return df
df_all = remove_one_hotness(df_all)

df_all.to_csv(f"{ANNOT_DIR}/master_annotations_origin.csv", index=False)
```

```python
# spreadsheet with only nih annotation
df_nih_only = df_all[df_all['origin'] == 'nih']
df_nih_only.to_csv(f"{ANNOT_DIR}/nih_annotations_origin.csv", index=False)
```

```python
# spreadsheet with all annotations except nih
df_no_nih = df_all[df_all['origin'] != 'nih']
df_no_nih.to_csv(f"{ANNOT_DIR}/no_nih_annotations_origin.csv", index=False)
```
